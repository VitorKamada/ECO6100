```python
import pandas as pd
file = "https://github.com/VitorKamada/ECO6100/raw/master/Data/AEJfigs.dta" 
df = pd.read_stata(file)
```

```python
df.head()
```

```python
df.describe()
```

```python
import numpy as np            
df['const'] = 1
df['age'] = df['agecell'] - 21
df['over21'] = np.where(df['agecell'] >=21, 1, 0)
df['over21_age'] = df['age']*df['over21']
df['age2'] = df['age']*df['age']
df['over21_age2'] = df['age2']*df['over21']
```

```python
df.head()
```

```python
import statsmodels.api as sm
result1 = sm.OLS(df['all'], df[['const', 'age', 'over21']],
                    missing='drop').fit()
print(result1.summary())
```

```python
result2 = sm.OLS(df['all'],
                    df[['const', 'age', 'over21', 'over21_age']],
                    missing='drop').fit()
print(result2.summary())
```

```python
result3 = sm.OLS(df['all'],
                    df[['const', 'age', 'age2', 'over21']],
                    missing='drop').fit()
print(result3.summary())
```

```python
result4 = sm.OLS(df['all'],
                    df[['const', 'age', 'age2', 
                        'over21', 'over21_age', 'over21_age2']],
                    missing='drop').fit()
print(result4.summary())
```

```python
df1 = df[df['agecell'] >= 20]
df1 = df1[df1['agecell'] <= 22]
result5 = sm.OLS(df1['mva'],
                    df1[['const', 'age', 'age2', 
                        'over21', 'over21_age', 'over21_age2']],
                    missing='drop').fit(cov_type='HC1')
print(result5.summary())
```
