```python
import pandas as pd
import statsmodels.api as sm
```

```python
file="https://github.com/VitorKamada/ECO6100/raw/master/Data/JTRAIN.DTA"
df = pd.read_stata(file)
df.head()
```

```python
df['const'] = 1
df = df[df['d88'] == 1]
reg1 = sm.OLS(df['lscrap'], df[['const', 'grant']], 
              missing='drop').fit()
print(reg1.summary())
```

```python
reg2 = sm.OLS(df['clscrap'], df[['const', 'cgrant']], 
              missing='drop').fit()
print(reg2.summary())
```

```python
import pandas as pd
import statsmodels.api as sm
file="https://github.com/VitorKamada/ECO6100/raw/master/Data/KIELMC.DTA"
df = pd.read_stata(file)
df.head()
```

```python
df['const'] = 1
df['rprice'] = df['rprice']/1000

df81 = df[df['year'] == 1981]
reg1 = sm.OLS(df81['rprice'], df81[['const', 'nearinc']],
                    missing='drop').fit()
print(reg1.summary())
```

```python
df78 = df[df['year'] == 1978]
reg2 = sm.OLS(df78['rprice'], df78[['const', 'nearinc']],
                    missing='drop').fit()
print(reg2.summary())
```

```python
reg3 = sm.OLS(df['rprice'], df[['const', 'y81', 'nearinc', 
              'y81nrinc']], missing='drop').fit()
print(reg3.summary())
```
